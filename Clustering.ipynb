```python
#using the elbow method to find number of clusters:
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, n_init = 10, random_state = 1).fit(X)
    inertias.append(km.inertia_)
```

```python
plt.plot(range(2, 11), inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

Elbow method suggests using 3 clusters - which would make sense seeing as there are 3 target labels. Now lets check with the silhouette score.

```python
from sklearn.metrics import silhouette_score
import sklearn.metrics as metrics

SK = range(2,12)
sil_score = []

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in SK:
    labels=KMeans(n_clusters=k, init="k-means++", n_init = 10, random_state=1,).fit(X).labels_
    score = metrics.silhouette_score(X,labels,metric="euclidean",sample_size=1000,random_state=200)
    sil_score.append(score)
    print ("Silhouette score for k(clusters) = "+str(k)+" is "
           +str(metrics.silhouette_score(X,labels,metric="euclidean",sample_size=1000,random_state=200)))
```

```python
sil_centers = pd.DataFrame({'Clusters' : SK, 'Sil Score' : sil_score})
sil_centers
```

```python
import seaborn as sns
sns.lineplot(x = 'Clusters', y = 'Sil Score', data = sil_centers, marker="+")
```

```python
from sklearn.metrics import balanced_accuracy_score
def balanced_accuracy_checker(y_true_inputted, y_pred_inputted):
    best_score = 0;
    best_encoding = None;
    encoding = None
    
    combinations = [[0,1,2], [0,2,1], [1,0,2], [1,2,0], [2,0,1], [2,1,0]]
    for combination in combinations:
        y_true_encoded = y_true_inputted
        y_true_encoded = y_true_encoded.replace(to_replace="Low risk", value = combination[0])
        y_true_encoded = y_true_encoded.replace(to_replace="Moderate risk", value = combination[1])
        y_true_encoded = y_true_encoded.replace(to_replace="High risk", value = combination[2])
        
        score = balanced_accuracy_score(y_true_encoded, y_pred_inputted)
        print(y_true_encoded)
        print(score)
        
        if(score > best_score):
            best_score = score
            best_encoding=y_true_encoded
            encoding = "Low risk: " + str(combination[0]) + "\nModerate risk: " + str(combination[1]) + "\nHigh risk: " + str(combination[2])
            
    print("\nBest balanced accuracy score: ",  best_score)
    print("Encoding: ", encoding)
```
